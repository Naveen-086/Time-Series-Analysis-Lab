```python
import matplotlib.pyplot as plt
```

scatter **plot**

```python
plt.scatter(dataset["USD (AM)"],dataset["GBP (AM)"])

plt.colorbar()
plt.show()
```

Bar **plot**

```python
plt.bar(dataset["USD (AM)"],dataset["GBP (AM)"])
```

**Histogram**

```python
plt.hist(dataset["USD (AM)"])
plt.hist(dataset["GBP (AM)"])
```

using **seaborn**

```python
import seaborn as sns
sns.lineplot(x="USD (AM)",y="GBP (AM)",data=dataset)
plt.title("using seaborn visualization")
```
